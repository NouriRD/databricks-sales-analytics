# Sales Analytics using Apache Spark & Delta Lake

## Project Overview
This project demonstrates a simple end-to-end data analytics pipeline using
Apache Spark on Databricks Community Edition.

### Objectives:
- Load and process sales data using Spark
- Perform transformations and aggregations
- Store data as a managed Delta Table
- Analyze sales using Spark SQL
- Visualize results using Python

## **Data Creation (Data Ingestion)**

```python
data = [
    (1, "Laptop", "Electronics", 1200, 2),
    (2, "Phone", "Electronics", 800, 3),
    (3, "Tablet", "Electronics", 500, 1),
    (4, "Desk", "Furniture", 300, 2),
    (5, "Chair", "Furniture", 150, 4)
]

columns = ["id", "product", "category", "price", "quantity"]

df = spark.createDataFrame(data, columns)
df.show()
```

## **Data Transformation**

```python
from pyspark.sql.functions import col

df = df.withColumn("total_sales", col("price") * col("quantity"))
df.show()
```

## **Store Data (Delta Lake â€“ Managed Table)**

```python
df.write \
  .format("delta") \
  .mode("overwrite") \
  .saveAsTable("sales_delta")
```

## **Data Analysis with Spark SQL**

```python
%sql
SELECT 
  category,
  SUM(total_sales) AS total_revenue
FROM sales_delta
GROUP BY category
```

## **Data Visualization**

```python
df_plot = (
    spark.table("sales_delta")
    .groupBy("category")
    .sum("total_sales")
    .toPandas()
)

import matplotlib.pyplot as plt

plt.figure()
plt.bar(df_plot["category"], df_plot["sum(total_sales)"])
plt.xlabel("Category")
plt.ylabel("Total Revenue")
plt.title("Total Sales by Category")
plt.show()
```

## **Results & Insights**

- Electronics category generates the highest revenue
- Furniture has lower revenue but higher quantity sold
- Spark SQL allows fast aggregation on structured data
- Delta Lake ensures reliable and structured storage
